```python
# turn "Customer lifetime Value" Column into a float by taking the the last lette % out

#x = value in the columns

def funct1(x):
# here we make sure it is a string
    x = str(x)
    
    if x != "nan":
# here we are selecting all the values of the string except for the last one -> which is the one we want to get rid of i.e. %
        y = x[: -1]
        return float(y)
    else:
        return 0

# here we are applyting the above defined function on any value in the customer_lifetime_value. First we are copying the df to the df2
df2 = df.copy()
df2["customer_lifetime_value"] = df2 ["customer_lifetime_value"].apply(funct1)
```
