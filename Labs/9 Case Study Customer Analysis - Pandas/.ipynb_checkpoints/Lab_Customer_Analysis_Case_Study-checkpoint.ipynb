```python
# turn "Number of Open Complaints" Column into a float -> by selecting the 3. letter of the value only

def funct2(x):
    x = str(x)
    
    if x.lower() == "nan":
        return 0
    else:
# here we are selecting only the 3. value of the string -> seems to be the only one showing the amount of open complaints
       return (int(x[2]))


# creating a new data frame and copying the df in order to apply function to the values of the column
df2 = df.copy()
df2["Number of Open Complaints"] = df2["Number of Open Complaints"].apply(funct2)
```

```python
# standardizing the Gender Column by changing its values to correct values

def gender_cl(x):


# why can you do it in this way to check for none value, because they have the value that when you equal none values, it results a false
# then I know that x is NaN

    if not x == x:
        return "n"

# filtering/overwritting

    else:
        dct_ = {"Male": "M",
           "female": "F" ,
           "Femal": "F" ,
            "F": "F" ,
            "M": "M"}
    
        return dct_[x]

#test it:
gender_cl("Femal")

```

```python
df2["Gender"] = df["Gender"].apply(gender_cl)

```

## Bucketing Columns
### Write a function to replace column "State" to different zones. California as West Region, Oregon as North West, and Washington as East, and Arizona and Nevada as Central

```python
# mapping the state names to the new column called region with different zones 

def region(x):
    
    if x !=x:
        return x

    else:
        dct_ = {"California": "West Region",
           "Washington": "East" ,
           "Oregon": "North West" ,
            "Arizona": "Central" ,
            "Nevada": "Central"}
    
        return dct_[x]
```

```python
region("California")
```

```python
df2["region"] = df2["State"].apply(region)
```
