## Standardizing data: Renaming Columns

```python
# list current column names

df.columns
```

```python
column_names = df.columns
column_names_clean = [column.lower().replace(" ", "_") for column in column_names]
df.columns = column_names_clean
df.info()
```

```python
df.head()
```

## Deleting Columns

```python
# delete the column "customer" as it is only a unique identifier for each row of data

df = df.drop(["customer"], axis =1) 

# axis=1 means column wise 
```

```python
# delete the column "unnamed:_0"

df = df.drop(["unnamed:_0"], axis =1) 

# axis=1 means column wise 
```
