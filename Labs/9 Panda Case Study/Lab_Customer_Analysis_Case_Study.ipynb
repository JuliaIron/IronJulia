## Setting up Panda 

```python
import pandas as pd
```

## Reading data 


```python
# the data is spread in two different .csv files

df1= pd.read_csv("file1.csv")
df1
```

```python
df2= pd.read_csv("file2.csv")
df2
```

```python
# in order to analyse the data, I concetinate the data from both files into a new file

# before that, I check if the columns of both files are identical.

set(df1.columns)== set(df2.columns)
```

## Concetinating files 

```python
pd.concat ([df1,df2], axis=0, ignore_index=True)

# axis=0 means row wise and is set as default, (axis=1 means column wise)
# ignore_index=True created a new index starting with 0

# alternative: 
# pd.concat([df1,df2]) 
```

```python
# getting info about the new data frame. 

# i.e. I see that there are floating data type and others (i.e. strings), called as objects
# i.e. I see that there are 7 null data cells in the customer lifetime value column 

df = pd.concat ([df1,df2], axis=0, ignore_index=True)

print(df.info())
```

```python
# the data frame exists of 5004 rows and 11 columns 

df.shape
```

## Standardizing data: Renaming Columns

```python
df.columns
```

```python
# changing name of column 2 from ST to State and of GENDER to Gender 
# with dictonary, old column names are keys, new column names are values

df.rename(columns={"ST":"State", "GENDER":"Gender"}, inplace=True)
```

```python
df.head()
```

## Deleting and Rearranging Columns

```python
# delete the column "customer" as it is only a unique identifier for each row of data

df = df.drop(["Customer"], axis =1) 

# axis=1 means column wise 

df.head()

#shows error, because I run the cell a second time. How to remove this? 
```
